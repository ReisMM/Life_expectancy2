# Life expectancy and GDP

## First analysis

```python
# Import libraries and data
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("all_data.csv")
data.columns = ['Country', 'Year', 'Life_exp', 'GDP']
print(data.info)
print(data.nunique())
```

In this dataset we can find the average Life expectancy at birth (years) and GDP for 6 countries, for 15 different years. 

```python
print(data.head(20))
```

See how may different countries and years do we have in our data

```python
countries = []
years = []

for i in data['Country']:
    if i not in countries:
        countries.append(i)
    else:
        continue

for a in data['Year']:
    if a not in years:
        years.append(a)
    else:
        continue

print(countries)
print(years)
```

## Basic graph Life expectancy and GDP

```python
plt.figure(figsize=(15,10))
ax=plt.subplot(2, 2, 1)
plt.plot(years, data.Life_exp[data.Country == 'Chile'])
plt.plot(years, data.Life_exp[data.Country == 'China'])
plt.plot(years, data.Life_exp[data.Country == 'Germany'])
plt.plot(years, data.Life_exp[data.Country == 'Mexico'])
plt.plot(years, data.Life_exp[data.Country == 'United States of America'])
plt.plot(years, data.Life_exp[data.Country == 'Zimbabwe'])
ax.legend(['Chile', 'China', 'Germany', 'Mexico', 'USA', 'Zimbabwe'])
plt.title('Life Expectancy vs years')
plt.xlabel('Years')
plt.ylabel('Life expectancy')


ax=plt.subplot(2, 2, 2)
plt.plot(years, data.GDP[data.Country == 'Chile'])
plt.plot(years, data.GDP[data.Country == 'China'])
plt.plot(years, data.GDP[data.Country == 'Germany'])
plt.plot(years, data.GDP[data.Country == 'Mexico'])
plt.plot(years, data.GDP[data.Country == 'United States of America'])
plt.plot(years, data.GDP[data.Country == 'Zimbabwe'])
ax.legend(['Chile', 'China', 'Germany', 'Mexico', 'USA', 'Zimbabwe'])
plt.title('GDP vs years')
plt.xlabel('Years')
plt.ylabel('GDP')

ax=plt.subplot(2, 2, 3)
plt.plot(years, data.GDP[data.Country == 'Chile'])
plt.plot(years, data.GDP[data.Country == 'Mexico'])
plt.plot(years, data.GDP[data.Country == 'Zimbabwe'])
ax.legend(['Chile', 'Mexico', 'Zimbabwe'])
plt.title('GDP vs years (detail)')
plt.xlabel('Years')
plt.ylabel('GDP')

plt.tight_layout()
plt.show()
plt.clf()
```

**Analysis:**

- Life expectancy: The country with the  highest life expectancy is Germany(max=81, avg=79.6), followed by Chile, USA, Mexico and China(max=76.1, avg= 74.3). In all this countries, life expectancy has consistantly increased over the years. Consequently, all these countries have reached the maximum of life expectancy in 2015 (last year of the study). The country with the lowest life expectancy is Zimbabwe however, from 2004 to 2015, was the country in which life expectancy has increase the most (from 44.3 to 60.7 years). 

- GDP: The country with highest GDP in 2015 is USA followed by China. These two countries had the biggest increase in GDP over 15 years: USA from 1.03e13 to 1.81e13 and China from 0.12e13 to 1.11e13. Next, we have Germany, Mexico, Chile and finally Zimbabwe which is the country with the lowest GDP. Those four countries, althought the GDP grows from 2000 t0 2015, do not show a clear increase in GDP.

## Relation between life expectancy and GDP

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x=data.GDP, y=data.Life_exp, hue=data.Country)
plt.title('Relation between life expectancy and GDP (per country)')
plt.show()
plt.clf()
```

**Analysis:**

GDP means Gross Domestic Product. A country with high GDP reflects a growing economy, with more resources available to the population. It would not be surprising to see that the population tend to live longer in countries with high GDP. In those countries it is expected that the population is vaccinated, work-life balance leading to better diet and exercise, and access to healthcare. The US from 2000 to 2015 only increase its life expectancy 2.5 years although the GDP increased 75%. This probably means that, after a certain point, the increase in the GDP does not produce an effect in life expectancy. 

In China, a growth in the GDP of 813% from 2000 to 2015 resulted in an increase in the life expectancy in 4 years. However we can see in the graph that the increase in the life expectancy is higher in the first 8 years with a slight increase in the GDP. This probably means that a slight increase in the GDP (193%) was sufficient to improve the living conditions of the chinese population. From 2007 to 2015 the GDP increased 211% but the life expectancy only increased 1.7 years.

The country with highest life expectancy is Germany although its GDP only increased 73% from 2000 to 2015. It's worth mentioning that its life expectancy was already high in the year 2000, but nevertheless increased 3 years in that time period. 

Chile is the second country with highest life expectancy but is one of the countries with lowest GDP. Over 15 years, its life expectancy increased 3 years, and the GDP 211%. It is worth to explore how the population of a country with such a low GDP has a high life expectancy. 

Finally, Zimbabwe had an increase of 143% in the GDP (althowh it's still the country with the lowest GDP), but the life expectancy increased 14.7 years to a maximum of 60.7 years in 2015 (which is still the country with the lowest life expectancy). Probably this means that an increase of 143% in the GDP is enough to dramatically change the life conditions of the population and thus people live longer (not as much as they could live in "more developed countries" with better life conditions). 

```python
def increase_years(countries):
    increase_y = []
    for i in countries:
        b = float(data.Life_exp[data.Country == i][data.Year == 2015])-float(data.Life_exp[data.Country == i][data.Year == 2000])
        increase_y.append(b)
    return increase_y

def increase_gdp(countries):
    increase_g = []
    for i in countries:
        c = (float((data.GDP[data.Country == i][data.Year == 2015])-float(data.GDP[data.Country == i][data.Year == 2000]))/(float(data.GDP[data.Country == i][data.Year == 2000])))*100
        increase_g.append(c)
    return increase_g


plt.figure(figsize=(15,3))
plt.plot(countries, increase_years(countries))
plt.title('Increase in life expectancy from 2000 to 2015')
plt.xlabel('Country')
plt.ylabel('Increase in life expectancy')
plt.show()
plt.clf()

plt.figure(figsize=(15,3))
plt.plot(countries, increase_gdp(countries))
plt.title('Increase in GDP from 2000 to 2015 (%)')
plt.xlabel('Country')
plt.ylabel('Increase in GDP (%)')
plt.show()
plt.clf()
```
